```python
from numpy import *
from matplotlib.pylab import *
```

```python
t = linspace(0,2*pi-(2*pi/32),32)

def par(t,w):
    return 3*cos(w*t) + 2*cos(3*w*t) + cos(5*w*t)

anpar = [3,2,1]
npar = [1,3,5] #para el resto es 0

def impar(t,w):
    return sin(w*t) + 2*sin(3*w*t) + 3*sin(5*w*t)

animpar = [1,2,3]
nimpar = [1,3,5]

def mixed(t,w):
    return 5*sin(w*t) + 2*cos(3*w*t) + sin(5*w*t)

anmix = [5,2,1]
nmix = [1,3,5]


def xk(lista):
    N=len(lista)
    lis = [0]*N
    for k in range(N):
        suma = 0
        for n in range(N):
            suma += exp(-2*pi*1j*n*k/N)*lista[n]
        lis[k]=abs(suma)
    return array(lis)/N

figure(figsize=(15,15))

subplot(3,2,1)
plot(t,par(t,1),'o')
title('par puntos') 
subplot(3,2,2)
title('par fourier') 
stem(xk(par(t,1)))


subplot(3,2,3)
plot(t,impar(t,1),'o')
title('impar puntos') 
subplot(3,2,4)
title('impar fourier') 
stem(xk(impar(t,1)))

subplot(3,2,5)
plot(t,mixed(t,1),'o')
title('mixed puntos') 
subplot(3,2,6)
title('mixed fourier')
stem(xk(mixed(t,1)))

savefig('1.png')
```

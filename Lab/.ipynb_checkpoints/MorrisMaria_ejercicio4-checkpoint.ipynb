```python
img3=img.copy()
img3[::]=0
x,y=np.meshgrid(range(len(img3)-500),range(len(img3[0])-250))
img3[x+200,y+50]=img[x+300,y+50]
imgplot = plt.imshow(img3)
```

```python
#Con ayuda de https://stackoverflow.com/questions/29330307/how-to-delete-a-set-of-meshgrid-points-inside-a-circle
img4=img.copy()
x0,y0,ra=500,450,230
x, y = np.meshgrid(range(len(img4)),range(len(img4)))
r = np.sqrt((x - x0)**2 + (y - y0)**2)
afuera = r>ra
for i in range(len(img4)):
    for j in range(len(img4[i])):
        if afuera[i][j]:
            img4[i][j]=0
imgplot = plt.imshow(img4)
```

```python
def intensity(r,g,b):
    return 0.299*r + 0.587*g + 0.114*b
for i in range(len(img4)):
    for j in range(len(img4[i])):
        a=img4[i][j]
        img4[i][j]=intensity(a[0],a[1],a[2])
imgplot = plt.imshow(img4,cmap='gray')
```

```python
from numpy import *
from numpy.linalg import *
from matplotlib.pylab import *
```

```python
archivo = open("datos.data",'r')
archivo = list(archivo)
archivo = array([i.split(',') for i in archivo])
def cambiar(clase):
    if clase == 'Iris-setosa\n':
        clase = 0
    elif clase == 'Iris-versicolor\n':
        clase = 1
    elif clase == 'Iris-virginica\n':
        clase = 2
    return clase

m = len(archivo)

sepal1 = array(list(map(float,array(archivo[:,0]).reshape(m,1))))
sepal1 = (sepal1 - sepal1.mean())/std(sepal1) 
sepal2 = array(list(map(float,array(archivo[:,1]).reshape(m,1))))
sepal2 = (sepal2 - sepal2.mean())/std(sepal2) 
petal1 = array(list(map(float,array(archivo[:,2]).reshape(m,1))))
petal1 = (petal1 - petal1.mean())/std(petal1) 
petal2 = array(list(map(float,array(archivo[:,3]).reshape(m,1))))
petal2 = (petal2 - petal2.mean())/std(petal2) 
tipo = array(list(map(cambiar,archivo[:,4])))
tipo = (tipo - tipo.mean())/std(tipo)
X = column_stack((sepal1,sepal2,petal1,petal2)) #no hace una gran diferencia en los datos incluir o no la categoria
figure()
hist(sepal1)
title('length sepal')
figure()
hist(sepal2)
title('wigth sepal')
figure()
hist(petal1)
title('lenght petal')
figure()
hist(petal2)
title('wigth petal')

covar = (1/(m-1))*X.T.dot(X)
u,s,v = svd(covar)

#para mantener la varianza suma hasta s_k debe ser del 90% de los s_n 
k = 0
vari = 0
while vari<0.9:
    k+=1
    vari = sum(s[:k])/sum(s)
    


ured = u[:,:2]
z = X.dot(ured)

figure()
hist(z[:,0])
title('componente 1')

figure()
hist(z[:,1])
title('componente 2')

figure()
scatter(z[:,0],z[:,1])
for i in range(4): 
    arrow(0,0,u[i][0],u[i][1],length_includes_head=True, head_width=0.08, color = 'darkorange')
    
print('Se puede evidenciar por la varianza de los datos que para las sepas la longitud y el ancho de los datos tiene una distribucion parecida a la normal. Sin embargo en el caso de los petalos hay dos grupos, por lo que uno podria decir que se pueden clasificar en dos grandes grupos. Por lo que se puede ver cuando se hacen las componentes principales que la varianza de los petalos permanece en la componente principal 1 y el de la sepa en la componente principal 2.')
```

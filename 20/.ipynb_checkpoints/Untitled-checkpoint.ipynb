```python
from numpy import *
from matplotlib.pylab import *
```

```python
h = 20
t = linspace(pi/2 - 2*pi,pi/2 -(pi/(2*h)),h)

s = lambda t: 1/(1-0.9*sin(t))

def xk(lista):
    N=len(lista)
    lis = [0]*N
    for k in range(N):
        suma = 0
        for n in range(N):
            suma += exp(-2*pi*1j*n*k/N)*lista[n]
        lis[k]=abs(suma)
    return array(lis)/N
h2 = 100
tau = linspace((pi/2 - 2*pi),pi/2 -(pi/(2*h)),h)
l = [cov(s(t+i))/var(s(t)) for i in t]

plot(t,s(t))
figure()
stem(xk(s(t)))
figure()
semilogy(t,xk(s(t))**2)
figure()
plot(t,l)
```
